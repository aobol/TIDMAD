## To plot your denoised file, replace `denoised_ABRA_limit_file` with the pathway to your CSV file outputted by the `brazilband.py` script.
If you wish to only plot the raw ABRA limit, simply assign `denoised_ABRA_limit_file = None`.
To choose the resulting figure name, change `fig_name`

```python
raw_ABRA_limit_file = "limit_data/TIDMAD_raw_limits.csv"
denoised_ABRA_limit_file = "limit_data/TIDMAD_denoised_FCNet_limits.csv"
fig_name = 'AxionPhoton'
```

```python
%matplotlib inline
from numpy import *
import matplotlib.pyplot as plt
from PlotFuncs_TIDMAD import FigSetup, AxionPhoton, MySaveFig, BlackHoleSpins, FilledLimit, line_background

fig, ax = FigSetup(Shape='Rectangular',ylab=r'$|g_{a\gamma\gamma}|$ [GeV$^{-1}$]',mathpazo=True)

# Other Limits
col = [120/255, 120/255, 120/255]
AxionPhoton.SHAFT_TIDMAD(ax,col=col,edgealpha=1,alpha=1)
col = [240/255, 240/255, 240/255]
AxionPhoton.LowMassAstroBounds_TIDMAD(ax,col=col)
AxionPhoton.QCDAxion_TIDMAD(ax)
col = [220/255, 220/255, 220/255]
AxionPhoton.HaloscopesUniform_TIDMAD(ax,text_on=False,col=col)
col = [200/255, 200/255, 200/255]
AxionPhoton.Helioscopes(ax,text_on=False,col=col)
ax.text(1e-4,1.5e-9,r'{\bf CAST}',fontsize=18,color='w',rotation=0,ha='center',va='top',clip_on=True,path_effects=line_background(1.5,'k'))
col = [160/255, 160/255, 160/255]
AxionPhoton.LSW_TIDMAD(ax, col=col)

# ABRA
AxionPhoton.ABRACADABRA_TIDMAD(ax, raw_ABRA_limit_file, denoised_fname=denoised_ABRA_limit_file, edgealpha=0)
AxionPhoton.ABRACADABRA_original(ax,col='pink',text_shift=[3,0.18])

plt.xlim(1e-11, 1e-3)
MySaveFig(fig,fig_name)
```
