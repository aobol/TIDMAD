```python
import numpy as np
import matplotlib.pyplot as plt
```

This plot aims to plot a tiny portion of the TIDMAD dataset for illustration purpose.

We curated two time series file in numpy format:
* `readout.npy` includes time series data from CH1: SQUID time series
* `injected.npy` includes time series data from CH2: inject time series
Note that these two time series are eactly the same length and one-to-one correspondent.

User can download these two files from [this link](https://drive.google.com/drive/folders/1vUrigYDlHU286d5e0aBcI8o101lm6r1P?usp=share_link)

In the following script, we attempted to plot the 10000th sample to the 101000th samples. Users can change `begin_sample` and `end_sample` to explore other parts of this dataset

```python
begin_sample = 10000
end_sample = begin_sample+ 1000

readout_data = np.load('readout.npy')[begin_sample:end_sample] # This is the CH1: SQUID time series [Noisy] 
injected_data = np.load('injected.npy')[begin_sample:end_sample] # This is the CH2: injected time series [Ground Truth] 

# Create x-axis values (assuming both arrays have the same length)
# If they have different lengths, you may need to adjust this
x = np.arange(len(readout_data))

# Create the plot
plt.figure(figsize=(10, 6))

# Plot both time series
plt.plot(x, readout_data, color='magenta', linewidth=2, label='Readout Signal')
plt.plot(x, injected_data, color='cyan', linewidth=2, label='Injected Signal')

# Add labels and title
plt.xlabel('Time Point')
plt.ylabel('Amplitude')
plt.title('Comparison of Readout and Injected Signals')

# Add legend
plt.legend()

# Add grid for better readability
plt.grid(True, alpha=0.3)

# Improve layout
plt.tight_layout()

# Show the plot
plt.show()
```
